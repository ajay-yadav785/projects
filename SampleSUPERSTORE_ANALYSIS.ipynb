```python
plt.figure(figsize=(6,6))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
print(df['State'].value_counts())
plt.figure(figsize=(17,15))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby('Region')[['Profit', 'Sales']].agg(['sum']).plot.bar() # Changed from tuple to list
plt.title('Total Profit Booked on Sale per Region')
plt.rcParams['figure.figsize'] = [10, 10]
plt.show()
```

```python
print(df['Sub-Category'].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
sub_category_profit = df.groupby('Sub-Category')['Profit'].sum()
print(sub_category_profit)
plt.figure(figsize=(10,10))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python

subcat_profit = df.groupby('Sub-Category')['Profit'].sum()

# Now you can plot the data
subcat_profit.plot(kind='bar')
plt.title('Total Profit by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```

```python
print(df['Region'].value_counts())
plt.figure(figsize=(10,8))
sns.countplot(x=df['Region'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x=df['Segment'])
```

```python

customer_segment = df.groupby('Segment').sum()
customer_segment["Profit"].plot(kind='pie', autopct='%1.1f%%', colors=['magenta', 'yellow', 'orange', 'red'])
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='Discount', y='Profit', data=df, color='red', label='Discount')
plt.show()
```

```python
print(df['Sales'].describe())
plt.figure(figsize=(10,8))
plt.grid()
sns.distplot(df['Sales'], color='b', bins=50 , hist_kws={'alpha': 0.5});
```

```python

city_profit = df.groupby('City')['Profit'].sum()

max_profit_city = city_profit.idxmax()
min_profit_city = city_profit.idxmin()
```

```python
city_profit = df.groupby('City')['Profit'].sum()


max_profit_city = city_profit.idxmax()
min_profit_city = city_profit.idxmin()

extreme_cities = city_profit.loc[[max_profit_city, min_profit_city]]
fig, ax = plt.subplots(figsize=(6,4))
extreme_cities.plot(kind='bar', color=['green', 'red'], ax=ax)
plt.title('Cities with Maximum and Minimum Profit')
plt.xlabel('City')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
