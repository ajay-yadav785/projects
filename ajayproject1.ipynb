```python
plt.figure(figsize=(10, 6))
sns.barplot(x='product_name', y='product_price', data=df.head(10), palette='viridis') # Changed df.head to df
plt.title('Product Price by Name')
plt.xlabel('Product_ Name')
plt.ylabel('Product_ Price')
plt.xticks(rotation=45)
plt.show()
```

```python
state_counts = df['customer_state'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette='viridis')
plt.title('Number of Orders by Customer State')
plt.xlabel('Customer State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
 df_max=df['product_name'].value_counts()
 print(df_max)
```

```python
product_names = [
    "Perfect Fitness Perfect Rip Deck",
    "LIJA Women's Argyle Golf Polo"
]
sales_counts = [2219, 1]
colors = ['green' if count == max(sales_counts) else 'red' for count in sales_counts]
plt.figure(figsize=(12, 8))
bars = plt.bar(product_names, sales_counts, color=colors)
plt.xlabel('Product')
plt.ylabel('Number of Sales')
plt.title('Number of Sales per Product')
plt.xticks(rotation=45, ha='right')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')

plt.tight_layout()
plt.xticks(rotation=0)
plt.show()
```

```python
bins = pd.cut(df['profit_per_order'], bins=20)
profit_counts = bins.value_counts().sort_index()
# Plot bar plot
plt.figure(figsize=(12, 8))
plt.bar(profit_counts.index.astype(str), profit_counts.values, color='skyblue', edgecolor='black')
# Customize the plot
plt.xlabel('Profit per Order')
plt.ylabel('Frequency')
plt.title('Distribution of Profit per Order')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
total_sales = df.groupby('shipping_mode')['sales_per_customer'].sum() # Changed data to df
total_sales.plot(kind='bar')
plt.xlabel('Shipping Mode')
plt.ylabel('Total Sales')
plt.title('Total Sales by Shipping Mode')
plt.show()
```

```python
status_counts = df['order_status'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Order Status Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='shipping_mode', y='profit_per_order', data=df)
plt.xlabel('Shipping Mode')
plt.ylabel('Average Profit per Order')
plt.title('Average Profit by Shipping Mode')
plt.show()
```

```python
payment_type_counts = df['payment_type'].value_counts()
colors = {
    'DEBIT': 'skyblue',
    'TRANSFER': 'lightgreen',
    'CASH': 'lightcoral',
    'PAYMENT': 'lightgoldenrodyellow'
}

plt.figure(figsize=(8, 8))
plt.pie(payment_type_counts, labels=payment_type_counts.index, autopct='%1.1f%%', colors=[colors[ptype] for ptype in payment_type_counts.index])
plt.title('Distribution of Payment Types')
plt.show()
```
